```python
# clean employee_comp data:
import pandas as pd

data = pd.read_csv("/Desktop/SJ Employee Comp/Employee_Comp_2025.csv")

# load column names

print(data.columns)
```

```python
# remove columns such as: _id and 'Defined Contribution Plan Contributions - City Paid' (i don't know what that is)
 
data = data.drop(['_id', 'Defined Contribution Plan Contributions - City Paid'], axis = 1)
data.head()
```

```python
# integer columns
columns_to_clean = [
    "Total Cash Compensation", "Base Pay", "Overtime", "Sick and Vacation Payouts",
    "Other Cash Compensation", "Medical Dental Vision", "Retirement Contributions (Normal Cost) - City Paid",
    "Long Term Disability, Life, Medicare", "Misc Employment Related Costs"
]

#remove the columns using lambda
data[columns_to_clean] = data[columns_to_clean].apply(lambda x: x.astype(str).str.replace(',', '', regex=True))

data.head(20)
```

```python
# any duplicates? 
print(data.duplicated().sum())

# nulls?
print(data.isnull().sum())

print(data['Total Cash Compensation'].isnull().sum())

# looking good
```
